```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df = pd.read_csv("/media/optimus/bravo/data science/DATABASE/Baby names/Baby names/names1881.csv", index_col=0)
```

```python
df.head()
```

```python
df.reset_index(level=0, inplace=True)
```

```python
df.head()
```

```python
df.columns=['name','gender','id']
df.head()
```

```python
del df['gender']
del df['id']
```

```python
df.head()
```

```python
text = " ".join(nm for nm in df.name)
stopwords = set(STOPWORDS)
stopwords.update(["alone"])
wordcloud = WordCloud(stopwords=stopwords,width=1366, height=768,max_words=1000).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
wordcloud.to_file("baby_names.png")
```

```python
L = ["Thanks You",
     "Once you replace negative thoughts with positive ones, you'll start having positive results.",
     "Are you sure",
     "The Way Get Started Is To Quit Talking And Begin Doing.",
     "The Pessimist Sees Difficulty In Every Opportunity. The Optimist Sees Opportunity In Every Difficulty"
    ]
#create new df 
df1 = pd.DataFrame({'col':L})
```
